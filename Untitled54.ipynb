```python
from transformers import AutoTokenizer
import pandas as pd

# Load tokenizer
tokenizer = AutoTokenizer.from_pretrained("bert-base-multilingual-cased")

# Ensure 'Arabic' and 'Tafseer' columns exist and are non-empty
required_columns = ['Arabic', 'Tafseer']
for col in required_columns:
    if col not in data.columns:
        raise ValueError(f"Column '{col}' is missing in the dataset.")
    if data[col].isnull().any():
        print(f"Warning: Column '{col}' contains null values. These will be replaced with empty strings.")
        data[col].fillna("", inplace=True)

# Tokenize Arabic and Tafseer text
try:
    data['Arabic_tokens'] = data['Arabic'].apply(
        lambda x: tokenizer.encode(x, add_special_tokens=True, truncation=True, max_length=512)
    )
    data['Tafseer_tokens'] = data['Tafseer'].apply(
        lambda x: tokenizer.encode(x, add_special_tokens=True, truncation=True, max_length=512)
    )
    print("Tokenization completed successfully!")
except Exception as e:
    print(f"Error during tokenization: {e}")

# Display a preview of the tokenized data
print(data[['Arabic', 'Arabic_tokens', 'Tafseer', 'Tafseer_tokens']].head())
```

```python
import pandas as pd
import numpy as np
import torch
from transformers import AutoTokenizer, AutoModel
from torch.utils.data import DataLoader, Dataset
from sklearn.metrics.pairwise import cosine_similarity

# Device configuration
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Load model and tokenizer
model_name = "bert-base-multilingual-cased"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModel.from_pretrained(model_name).to(device)

# Dataset class for batch processing
class TextDataset(Dataset):
    def __init__(self, texts, tokenizer, max_length=512):
        self.texts = texts
        self.tokenizer = tokenizer
        self.max_length = max_length

    def __len__(self):
        return len(self.texts)

    def __getitem__(self, idx):
        text = self.texts[idx]
        try:
            encoded = self.tokenizer(
                text,
                truncation=True,
                padding="max_length",
                max_length=self.max_length,
                return_tensors="pt",
            )
            return {key: val.squeeze(0) for key, val in encoded.items()}
        except Exception as e:
            print(f"Error encoding text at index {idx}: {e}")
            return None

# Function to compute embeddings in batches
def compute_embeddings(dataloader, model, device):
    embeddings = []
    model.eval()
    for batch in dataloader:
        inputs = {key: val.to(device) for key, val in batch.items() if val is not None}
        with torch.no_grad():
            outputs = model(**inputs)
        batch_embeddings = outputs.last_hidden_state.mean(dim=1).cpu().numpy()
        embeddings.extend(batch_embeddings)
    return np.array(embeddings)

# Function to compute embeddings for a query
def get_query_embedding(query, tokenizer, model, device):
    try:
        inputs = tokenizer(query, return_tensors="pt", truncation=True, padding=True, max_length=512).to(device)
        with torch.no_grad():
            outputs = model(**inputs)
        return outputs.last_hidden_state.mean(dim=1).cpu().numpy()
    except Exception as e:
        print(f"Error computing query embedding: {e}")
        return None

# Function to perform semantic search
def semantic_search(query, data, embeddings, top_k=5):
    query_embedding = get_query_embedding(query, tokenizer, model, device)
    if query_embedding is None:
        print("Failed to compute query embedding. Aborting search.")
        return None, None

    try:
        similarities = cosine_similarity(query_embedding, embeddings)
        top_indices = np.argsort(similarities[0])[::-1][:top_k]
        results = data.iloc[top_indices]
        scores = similarities[0][top_indices]
        return results, scores
    except Exception as e:
        print(f"Error during semantic search: {e}")
        return None, None

# Load dataset
try:
    data = df # Replace with your dataset path
    assert "Arabic" in data.columns and "Tafseer" in data.columns, "Required columns missing in dataset."
except Exception as e:
    raise ValueError(f"Error loading dataset: {e}")

# Prepare datasets and dataloaders
batch_size = 32
arabic_texts = data['Arabic'].dropna().tolist()
tafseer_texts = data['Tafseer'].dropna().tolist()

arabic_dataset = TextDataset(arabic_texts, tokenizer)
tafseer_dataset = TextDataset(tafseer_texts, tokenizer)

arabic_loader = DataLoader(arabic_dataset, batch_size=batch_size, shuffle=False)
tafseer_loader = DataLoader(tafseer_dataset, batch_size=batch_size, shuffle=False)

# Check or generate embeddings
try:
    arabic_embeddings = np.load("arabic_embeddings.npy")
    tafseer_embeddings = np.load("tafseer_embeddings.npy")
    print("Loaded pre-computed embeddings.")
except FileNotFoundError:
    print("Generating embeddings...")
    arabic_embeddings = compute_embeddings(arabic_loader, model, device)
    tafseer_embeddings = compute_embeddings(tafseer_loader, model, device)

    # Save embeddings for future use
    np.save("arabic_embeddings.npy", arabic_embeddings)
    np.save("tafseer_embeddings.npy", tafseer_embeddings)

# Add embeddings back to the DataFrame
data['Arabic_embeddings'] = list(arabic_embeddings)
data['Tafseer_embeddings'] = list(tafseer_embeddings)

# Example query
query = "day of judgement"

# Perform search on Arabic text
arabic_results, arabic_scores = semantic_search(query, data, np.vstack(data['Arabic_embeddings']), top_k=5)

# Perform search on Tafseer text
tafseer_results, tafseer_scores = semantic_search(query, data, np.vstack(data['Tafseer_embeddings']), top_k=5)

# Display results
if arabic_results is not None:
    print("Top Results from Arabic Text:")
    for i, (text, score) in enumerate(zip(arabic_results['Arabic'], arabic_scores)):
        print(f"{i+1}. {text} (Score: {score:.4f})")

if tafseer_results is not None:
    print("\nTop Results from Tafseer:")
    for i, (text, score) in enumerate(zip(tafseer_results['Tafseer'], tafseer_scores)):
        print(f"{i+1}. {text} (Score: {score:.4f})")
```

```python
from sklearn.metrics.pairwise import cosine_similarity, euclidean_distances
import numpy as np
import torch

# Function to compute embeddings for a query
def get_query_embedding(query, tokenizer, model, device):
    inputs = tokenizer(query, return_tensors="pt", truncation=True, padding=True, max_length=512).to(device)
    with torch.no_grad():
        outputs = model(**inputs)
    return outputs.last_hidden_state.mean(dim=1).cpu().numpy()

# Function to compute various similarity metrics
def compute_similarity_metrics(query_embedding, embeddings):
    """
    Compute similarity scores using multiple metrics.
    Args:
        query_embedding (numpy.ndarray): Query embedding.
        embeddings (numpy.ndarray): Precomputed embeddings for the dataset.

    Returns:
        dict: Dictionary containing similarity scores for each metric.
    """
    similarities = {
        "cosine_similarity": cosine_similarity(query_embedding, embeddings)[0],
        "euclidean_distance": -euclidean_distances(query_embedding, embeddings)[0],  # Negated for ranking
        "manhattan_distance": -np.sum(np.abs(query_embedding - embeddings), axis=1),  # Negated for ranking
        "dot_product": np.dot(query_embedding, embeddings.T)[0],
    }
    return similarities

# Function to perform semantic search with all metrics
def semantic_search_all_metrics(query, data, embeddings, top_k=5):
    """
    Perform semantic search using multiple metrics.
    Args:
        query (str): Input query text.
        data (DataFrame): Dataset containing the text.
        embeddings (numpy.ndarray): Precomputed embeddings for the dataset.
        top_k (int): Number of top results to return.

    Returns:
        dict: Top-k results and scores for each metric.
    """
    query_embedding = get_query_embedding(query, tokenizer, model, device)
    similarity_scores = compute_similarity_metrics(query_embedding, embeddings)

    results = {}
    for metric, scores in similarity_scores.items():
        top_indices = np.argsort(scores)[::-1][:top_k]
        results[metric] = {
            "texts": data.iloc[top_indices],
            "scores": scores[top_indices],
        }
    return results

# Example query
query = "day of judgement"

# Perform search on Arabic text
try:
    arabic_results = semantic_search_all_metrics(query, data, np.vstack(data['Arabic_embeddings']), top_k=5)
    print("Top Results from Arabic Text:")
    for metric, result in arabic_results.items():
        print(f"\nMetric: {metric}")
        for i, (text, score) in enumerate(zip(result['texts']['Arabic'], result['scores'])):
            print(f"{i+1}. {text} (Score: {score:.4f})")
except Exception as e:
    print(f"Error performing semantic search on Arabic text: {e}")

# Perform search on Tafseer text
try:
    tafseer_results = semantic_search_all_metrics(query, data, np.vstack(data['Tafseer_embeddings']), top_k=5)
    print("\nTop Results from Tafseer:")
    for metric, result in tafseer_results.items():
        print(f"\nMetric: {metric}")
        for i, (text, score) in enumerate(zip(result['texts']['Tafseer'], result['scores'])):
            print(f"{i+1}. {text} (Score: {score:.4f})")
except Exception as e:
    print(f"Error performing semantic search on Tafseer text: {e}")
```

```python
# Input Section for Custom Predictions
def predict():
    print("\n=== Custom Prediction Section ===")
    query = input("Enter your query (e.g., 'day of judgement'): ").strip()

    if not query:
        print("Query cannot be empty. Please try again.")
        return

    # Perform search on Arabic text
    print("\nSearching Arabic Text...")
    arabic_results, arabic_scores = semantic_search(query, data, np.vstack(data['Arabic_embeddings']), top_k=5)

    # Perform search on Tafseer text
    print("\nSearching Tafseer Text...")
    tafseer_results, tafseer_scores = semantic_search(query, data, np.vstack(data['Tafseer_embeddings']), top_k=5)

    # Output Section for Predictions
    print("\n=== Prediction Results ===")

    # Determine the best results from Arabic and Tafseer
    if arabic_results is not None and tafseer_results is not None:
        best_arabic = arabic_results.iloc[0]['Arabic']
        best_tafseer = tafseer_results.iloc[0]['Tafseer']
        print(f"Best Result: {best_arabic}, {best_tafseer}")
    else:
        print("No results found. Please check your query or dataset.")

    # Display additional results for context
    print("\nTop Results from Arabic Text:")
    for i, (text, score) in enumerate(zip(arabic_results['Arabic'], arabic_scores)):
        print(f"{i+1}. {text} (Score: {score:.4f})")

    print("\nTop Results from Tafseer:")
    for i, (text, score) in enumerate(zip(tafseer_results['Tafseer'], tafseer_scores)):
        print(f"{i+1}. {text} (Score: {score:.4f})")

    print("\nEnd of Prediction.")

# Run Prediction Loop
if __name__ == "__main__":
    while True:
        print("\n=== Semantic Search Application ===")
        print("1. Enter a query for semantic search")
        print("2. Exit")
        choice = input("Choose an option (1/2): ").strip()

        if choice == "1":
            predict()
        elif choice == "2":
            print("Exiting application. Goodbye!")
            break
        else:
            print("Invalid choice. Please select 1 or 2.")
```
